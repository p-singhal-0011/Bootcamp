```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Q2import csv file diwali sales data set

```python
df=pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
```

##DROP UNRELATED/BLANK COLUMNS 'STATUS' AND 'UNNAMED1'

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

##CHECK FOR NULL VALUES AND FIND WHICH COLUMNS HAVE NULL VALUES

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

##Calculate the mean value of age , orders and amount by using describe function

```python
print(df.describe())
```

##PLOT A BAR CHART FOR GENDER AND ITS COUNT

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

##Plot a bar chart for gender vs total amount

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='Gender', y='Amount', data=df, estimator=sum , palette = 'rocket')
plt.title('Total Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```
