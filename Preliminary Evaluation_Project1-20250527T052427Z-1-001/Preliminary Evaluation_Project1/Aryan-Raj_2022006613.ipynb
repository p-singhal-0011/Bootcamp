## Q9. Plot a bar chart for Gender and it's count 

```python
data['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen'])

plt.title('Count of Each Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```
