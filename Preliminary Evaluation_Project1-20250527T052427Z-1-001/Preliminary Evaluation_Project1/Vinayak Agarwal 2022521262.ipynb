## Q9. Plot a bar chart for Gender and it's count 

```python
count=data["Gender"].value_counts()
print(count)
```

```python
plt.bar(count.index,count.values)
plt.xlabel("count")
plt.ylabel("gender")
plt.title("Gender vs Count")
plt.show()
```

## Q11. Plot the countplot for Age Group with hue and Gender

```python
import seaborn as sns
sns.countplot(data=data,x='Age Group',hue='Gender')
plt.title("Countplot age group vs gender")
plt.xlabel("Count")
plt.ylabel("Age Group")
plt.show()
```

## Q12. Draw the plot for total number of orders from top 10 states 

```python
top_state=data["State"].value_counts().head(10)
plt.figure(figsize=(18,6))
plt.bar(top_state.index,top_state.values)
plt.title("Number of orders vs top 10 states")
plt.xlabel("State")
plt.ylabel("Number of orders")
plt.show()
```

## Q13.Draw the plot to plot total amount/sales from top 10 states

```python
top_state_sales=data.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(18,6))
plt.bar(top_state_sales.index,top_state_sales.values)
plt.ylabel("Sales")
plt.xlabel("state")
plt.title("Amount of sales in each state")
plt.show()
```

## Q14. Draw the plot to show the marital status 

```python
data['Marital_Status']=data['Marital_Status'].map({0:'Unmarried',1:'married'})
martial_status=data["Marital_Status"].value_counts()
plt.bar(martial_status.index,martial_status.values)
```

## Q15. Draw the plot to show the top 10 occupations

```python
top_occ=data["Occupation"].value_counts().head(10)
plt.figure(figsize=(18,6))
plt.bar(top_occ.index,top_occ.values)
plt.title("Top 10 occutpation")
plt.xlabel("Occupation")
plt.ylabel("count")
plt.show()
```

## Q16. Draw the countplot to print the top 10 categories 

```python
top_cat=data["Product_Category"].value_counts().head(10)
plt.figure(figsize=(18,6))
plt.bar(top_cat.index,top_cat.values)
plt.title("Top 10 Category")
plt.xlabel("Category")
plt.ylabel("count")
plt.show()
```

## Q17. What is the prodct ID of the top 10 most sold products.

```python
top_pid=data["Product_ID"].value_counts().head(10)
plt.figure(figsize=(18,6))
plt.bar(top_pid.index,top_pid.values)
plt.title("Top 10 product id")
plt.xlabel("Product id")
plt.ylabel("count")
plt.show()
```
